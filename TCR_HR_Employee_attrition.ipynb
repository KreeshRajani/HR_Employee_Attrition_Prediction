## Some columns in the data set are not needed for training the machine learning model like employee number (it is just number identification ). 
## Similarly colums like standard hours employeeCount and Over18

```python
columns_drop = ['EmployeeNumber','EmployeeCount','Over18','StandardHours']
data = data.drop(columns_drop, axis=1)
```

## Transform non-numeric columns into numeric columns

```python
from sklearn.preprocessing import LabelEncoder
x = data.drop('Attrition', axis=1)
y = data['Attrition']
for column in x.columns:
    if x[column].dtype == np.number:
        continue
    x[column] = LabelEncoder().fit_transform(x[column])
```

```python
x['Age_Years'] = x['Age']
x = x.drop('Age', axis=1)
x
```

```python
print(x.columns)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.25, random_state = 0)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
forest.fit(X_train,Y_train)
```

```python
score = forest.score(X_train, Y_train)
print('Accuracy Score:', np.abs(score)*100)
```

### Accuracy Score: 99.5%

```python
predict = forest.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import confusion_matrix, classification_report
M = confusion_matrix(Y_test, predict)
TN = M[0][0]
TP = M[1][1]
FN = M[1][0]
FP = M[0][1]

print(M)
print('Model testing accuracy="{}!"'.format((TP+TN)/(TP+TN+FN+FP)))
print(classification_report(Y_test, predict))
```
